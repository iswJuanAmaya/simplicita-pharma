```python
import pandas as pd
import seaborn as sns
```

```python
imports = pd.read_csv('../match/perfect_match_v3.csv', encoding="ISO-8859-1")
```

```python
imports['nombre_medicamento'] = imports.apply(lambda row: row['nombre'] + ' ' +
                          row['marca'] + ' ' +
                          str(row['presentacion']) + ' ' +
                          row['forma_farma'] + ' ' +
                          str(row['concentracion']) + ' ' +
                          row['unidad_concentracion'], axis=1)
```

```python
imports.keys()
```

```python
imports.groupby('nombre_medicamento').agg({'valor_unitario':['count','min','max','mean','median',list]}).to_csv('dist_precios.csv')
```

```python
imports.groupby('nombre_medicamento').agg({'valor_unitario':["count", list]}).to_csv('prueba.csv')
```

```python
abemaciclib = imports[imports['nombre_medicamento'] == 'abemaciclib verzenio 56 tabletas 100.0 mg']
```

```python
sns.histplot(abemaciclib, x='valor_unitario', hue='Fecha')
```
